```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x = np.linspace(1,11,20)
y= randint(1,51,20)
y = np.sort(y)
```

```
plt.plot(x,y,c='m',marker='*',markersize=10,linestyle='--',lw=2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y')
```

```
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Subplot1')
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('Sublot2')
```

Object oriented

```
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('X vs y' )
```

2 fig in 1 frame

```
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.3,0.3,0.4,0.3])
ax1.plot(x,y,c='b')
ax2.plot(x,y,c='r')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_title('Plot1')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Plot2')
```

```
fig , axes = plt.subplots(1,3)
plt.tight_layout()
for curr_ax in axes:
    curr_ax.plot(x,y)

```

```
fig , axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(x,y)
axes[0].set_xlabel('X axis')
axes[1].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')
axes[1].set_ylabel('Y axis')
axes[0].set_title('Plot 0')
axes[1].set_title('Plot 1')
```

Graphs

```
y_views = [123,435,754,890,145,234,337]
i_views = [256,454,772,875,998,289,754]
t_views = [221,342,457,789,332,557,887]
days = [1,2,3,4,5,6,7]
```

```
plt.scatter(days,y_views,label='Youtube Views',marker='*')
plt.scatter(days,i_views,label='Insta Views',marker='*')
plt.scatter(days,t_views,label='Twitter Views',marker='*')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media Views')
plt.legend()
plt.show()
```
